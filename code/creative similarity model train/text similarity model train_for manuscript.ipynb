```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch
import time
import pickle
from tqdm import tqdm
```

```python
mydir = '' # local data source directory
data_path = '' # my file path
```

```python
df = pd.read_csv(data_path)
```

```python
''''
text fields for dedup from the original text table
select_cols = ['advertiser_name', 'ad_title', 'ad_text',
               'google_asr_text', 'aws_ocr_img_text', 'aws_ocr_video_text','description', 'checksum']
''''
```

```python
unique = df.drop('ad_id', axis=1).drop_duplicates(subset=['txt'], keep='last').reset_index(drop=True)
```

```python
data = unique.txt.tolist()
len(data)
```

## Train

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embedder
```

```python
# Start the timer
start_time = time.time()

# Your script or code here
corpus_embeddings = embedder.encode(data, convert_to_tensor=True)

# End the timer
end_time = time.time()

# Calculate the execution time
execution_time = end_time - start_time

# Print the execution time
print("Execution time:", execution_time, "seconds")
```
